```python
from pyspark.sql.functions import regexp_replace, concat_ws, col, expr, trim,right,substring

col_list = ["Name", "LastName", "MiddleName"]
pattern  = "[^a-zA-Z0-9]"
df = (df
    .withColumn("Name",regexp_replace(col("Name"),pattern, ""))
    .withColumn("LastName",regexp_replace(col("LastName"),pattern, ""))
    .withColumn("MiddleName",regexp_replace(col("MiddleName"), pattern, ""))
    .fillna("", subset=col_list)
    .withColumn("FullName", concat_ws(" ", *col_list))
    .withColumn("SalesPerson",
        regexp_replace(substring(col("SalesPerson"),18,100),
                       pattern, ""))
    .withColumn("Contact",
        regexp_replace(col("Phone"), r"[-()\s]", ""))
    .drop(*col_list, "Suffix", "Phone")
)

df.display()
```
